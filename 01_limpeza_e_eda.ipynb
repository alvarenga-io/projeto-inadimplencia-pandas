##1. Bibliotecas

```python
import pandas as pd
```

##2. Carregando Dados

```python
df = pd.read_excel('https://github.com/TrainingByPackt/Data-Science-Projects-with-Python/raw/refs/heads/master/Data/default_of_credit_card_clients__courseware_version_1_21_19.xls')
```

```python
print(df.columns)
len(df.columns)
```

###3.3 Verificando se os IDs são únicos

```python
df['ID'].nunique()
```

###3.5 Contagem de IDs

```python
id_counts = df['ID'].value_counts()
id_counts.head()
```

###3.6 Contagem de Duplicatas

```python
id_counts.value_counts()
```

##4. Análise das Duplicatas

###4.1 IDs que se repetem

```python
dupe_mask = id_counts == 2
dupe_mask
```

###4.2 Selecionar os IDs duplicados

```python
dupe_ids = id_counts.index[dupe_mask]
dupe_ids
```

###4.3 Transformar o index em lista

```python
dupe_ids = list(dupe_ids)
len(dupe_ids)
```

###4.4 Exibição

```python
dupe_ids[:5]
```

###4.5 Selecionando linhas com IDS duplicados

```python
df.loc[df['ID'].isin(dupe_ids), :].tail(10)
```

###4.6 Preparando a matriz booleana para filtragem do datafram


```python
df_zero_mask = df == 0
df_zero_mask
```

###4.7 Criando uma série booleana

```python
feature_zero_mask = df_zero_mask.iloc[:,1:].all(axis=1)
feature_zero_mask.head()
```

###4.8 Visualizar linhas zeradas

```python
zero_lines = df[feature_zero_mask]
zero_lines
```

###4.9 Eliminar duplicatas

```python
df_cleaned = df.loc[~feature_zero_mask,:].copy()
```

###4.10 Verificar shape do Dataframe


```python
df_cleaned.shape
```

```python
df_cleaned['ID'].nunique()
```

##5. Exploração e Limpeza de Dados

###5.1 Informações do Dataset

---



```python
df_cleaned.reset_index(drop=True).info()
```

###5.2 Investigando valores da coluna PAY_1





```python
df_cleaned['PAY_1'].value_counts()
```

###5.3 Criar mascara booleana para remover valores ausentes








```python
valid_pay1_mask = df_cleaned['PAY_1'] != 'Not available'
valid_pay1_mask[0:5]
```

```python
sum(valid_pay1_mask)
```

###5.4 Eliminar valores ausente

```python
df_valid = df_cleaned.loc[valid_pay1_mask,:].copy()
df_valid
```

###5.5 Verificar estrtutura do novo dataset

```python
df_valid.shape
```

```python
df_valid.reset_index(drop=True).info()
```

###5.6 Reatribuição do tipo de dado a coluna `PAY_1`




```python
df_valid['PAY_1'] = df_valid['PAY_1'].astype(int)
```

```python
df_valid[['PAY_1', 'PAY_2']].info()
```

**Explicações adicionais sobre as colunas feitas pelo sócio da empresa**


*   -2: significa que iniciou o mês sem valor pendente e o crédito não foi usado
*   -1: significa que usou um valor que foi totalmente pago
*   0: significa que o pagamento mínimo foi feito, mas o saldo devedor total não foi pago.
